```python
import pandas as pd
from functions import parameters as p


proj_param = p.Project_Parameters()

# Functions for cleaning geo for merging
def treat_geo_Canada(x):
    if x.strip() == 'Canada':
        x = 'Canada [00]'
    
    return x


def split_geo(x):
    val1 = x.split(' [')[0]
    val2 = x.split(' [')[1].replace(']', '')
    return [val1, val2]


def get_geo_name(x):
    return split_geo(x)[0]


def get_geo_code(x):
    return split_geo(x)[1]
```

```python
csi_path = proj_param.data_folder + "/crime_severity.parquet"
df_csi = pd.read_parquet(csi_path)
df_csi['geo'] = df_csi['geo'].apply(treat_geo_Canada)
df_csi['geo_name'] = df_csi['geo'].apply(get_geo_name)
df_csi['geo_code'] = df_csi['geo'].apply(get_geo_code)
display(df_csi.head(10))
print(df_csi.shape)
```

```python
ivd_path = proj_param.data_folder + "/incident_violation_detail.parquet"
df_ivd = pd.read_parquet(ivd_path)
df_ivd['geo'] = df_ivd['geo'].apply(treat_geo_Canada)
df_ivd['geo_name'] = df_ivd['geo'].apply(get_geo_name)
df_ivd['geo_code'] = df_ivd['geo'].apply(get_geo_code)
display(df_ivd.head(10))
print(df_ivd.shape)
```

```python
merged_data = (pd.merge(df_csi, df_ivd, how = "outer" , on = ['ref_date','geo_code'])
               .reset_index(drop=True)
               .drop(columns=['geo_x', 'geo_name_y', 'geo_y'])
               .rename(columns={'geo_name_x': 'geo_name'})
)
display(merged_data.head(50))
print(merged_data.shape)
```
